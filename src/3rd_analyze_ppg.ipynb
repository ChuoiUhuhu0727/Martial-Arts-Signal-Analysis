```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load file CSV
# Đảm bảo file "aikido_data_final.csv" nằm cùng thư mục với file notebook này
file_path = r"C:\Users\Admin\Downloads\draft_one_signal_final_project\aikido_data_final.csv"
df = pd.read_csv(file_path)

# 2. Xử lý Timestamp (Computational Thinking)
# Chuyển đổi Unix Timestamp sang "Elapsed Time" (giây) tính từ 0 để dễ nhìn
df['Time_Seconds'] = df['Timestamp'] - df['Timestamp'].iloc[0]

# 3. Visualize Raw Data
plt.figure(figsize=(15, 6))

# Vẽ đồ thị: X là thời gian, Y là tín hiệu Heart_IR
plt.plot(df['Time_Seconds'], df['Heart_IR'], color='magenta', label='Raw PPG Signal')

# Thêm Title và Label (Critical Thinking: Luôn phải có đơn vị)
plt.title("Visualizing Raw PPG Signal from MAX30102", fontsize=14)
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Raw IR Intensity", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

# 4. Zoom vào một khoảng ngắn (ví dụ 5 giây đầu) để thấy nhịp tim rõ hơn
plt.figure(figsize=(15, 4))
short_df = df[df['Time_Seconds'] < 5] 
plt.plot(short_df['Time_Seconds'], short_df['Heart_IR'], color='blue', label='Zoom-in (First 5s)')
plt.title("Detailed View: Identifying Heart Beats", fontsize=14)
plt.xlabel("Time (seconds)")
plt.ylabel("IR Intensity")
plt.grid(True)
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Load file CSV
file_path = r"C:\Users\Admin\Downloads\draft_one_signal_final_project\3rd_raw_data.csv"
df = pd.read_csv(file_path)

# --- BƯỚC QUAN TRỌNG: CHUYỂN ĐỔI SANG DẠNG SỐ ---
# errors='coerce' sẽ giúp chuyển các dòng lỗi (nếu có) thành NaN để tránh sập code
df['Timestamp'] = pd.to_numeric(df['Timestamp'], errors='coerce')
df['Heart_IR'] = pd.to_numeric(df['Heart_IR'], errors='coerce')

# Loại bỏ các dòng bị trống (NaN) sau khi chuyển đổi
df = df.dropna()

# 2. Xử lý Timestamp (Tính thời gian trôi qua)
# Bây giờ phép trừ này sẽ hoạt động vì cả hai đều là số (float)
df['Time_Seconds'] = df['Timestamp'] - df['Timestamp'].iloc[0]

# 3. Visualize Raw Data
plt.figure(figsize=(15, 6))
plt.plot(df['Time_Seconds'], df['Heart_IR'], color='magenta', label='Raw PPG Signal')
plt.title("Visualizing Raw PPG Signal from MAX30102", fontsize=14)
plt.xlabel("Time (seconds)", fontsize=12)
plt.ylabel("Raw IR Intensity", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend()

plt.show()
```
